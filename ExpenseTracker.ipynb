```python
class Expense:
    def __init__(self,category,amount,date):
        self.category = category
        self.amount = amount
        self.date = date

class ExpenseTracker:
    def __init__(self):
        self.expenses = []
    
    def add_expense(self):
        category = input("Enter expense category: ")
        amount = float(input("Enter expense price: "))
        date = input("Enter date in form of DD-MM-YYYY: ")

        expense = Expense(category,amount,date)
        self.expenses.append(expense)
        print("Expense added successfully!")

    def view_expense(self):
        if not self.expenses:
            print("No expense to show!")
            return
            
        for expense in self.expenses:
            print(f"Category: {expense.category}\nAmount: {expense.amount}\nDate: {expense.date}")

        print("_____________________________________")

    def delete_expense(self):
        if not self.expenses:
            print("No expense to delete.")
            return

        self.view_expense()
        try:
            index = int(input("Enter the index of expense you want to delete: "))
            index = index - 1

            if 0 <= index <= len(self.expenses):
                deleted = self.expenses.pop(index)
                print(f"Deleted: {deleted.category} - {deleted.amount}")
            else:
                print("Wrong Index!")

        except ValueError:
            print("Enter only number!")

    def total_expense(self):
        total = sum(expense.amount for expense in self.expenses)        
        print("Total Expense: ",total)

# Min Program
tracker = ExpenseTracker()

while True:
    print("========Expense Tracker========")
    print("1. Add Expense")
    print("2. View Expense")
    print("3. Delete Expense")
    print("4. Total Expense")
    print("5. Exit")

    try:
        choice = int(input("Enter your choice: "))
    except ValueError:
        print("Invalid Choice! Please correct number.")
        continue
    
    if choice == 1:
        tracker.add_expense()
    
    elif choice == 2:
        tracker.view_expense()

    elif choice == 3:
        tracker.delete_expense()

    elif choice == 4:
        tracker.total_expense()

    elif choice == 5:
        print("You are Exiting!")
        break         # To stop the loop

    else:
        print("Invalid Choice!")

#if __name__ == "__main__":
 #   main()
        
```
